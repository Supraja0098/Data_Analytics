```python
bigrams_data = """The mischievous younger sister pranked the sleepy older sister
the sleepy older sister got her revenge by pranking younger sister with a fake spider
in the End both sister pranked 
"""

bigrams_data_file = "bigrams_data.txt"

with open(bigrams_data_file, "w") as file:
    file.write(bigrams_data)
```

```python
from collections import defaultdict

def mapper(bigrams_data):
    word_counts = defaultdict(int)
    for line in bigrams_data:
        words = line.strip().split()
        for i in range(len(words) - 1):
            bigram = (words[i], words[i+1])
            word_counts[bigram] += 1
    print("Intermediate Pairs:")
    for bigram, count in word_counts.items():
        print(f"{bigram[0]}, {bigram[1]}: {count}")
    return word_counts.items()

def reducer(intermediate_pairs):
    bigram_counts = defaultdict(int)
    for bigram, count in intermediate_pairs:
        bigram_counts[bigram] += count
    print("\nFinal Output:")
    for bigram, count in bigram_counts.items():
        print(f"{bigram[0]}, {bigram[1]}: {count}")
    return bigram_counts.items()

# Read test data from file
bigrams_data_file = "bigrams_data.txt"
with open(bigrams_data_file, "r") as file:
    bigrams_data = file.readlines()

print("Input Data:")
for line in bigrams_data:
    print(line.strip())

# Call the mapper function with the input data
intermediate_pairs = mapper(bigrams_data)

# Sort the intermediate pairs based on keys
sorted_pairs = sorted(intermediate_pairs)

# Call the reducer function with sorted pairs
bigrams_output = reducer(sorted_pairs)

# Save the final output to a new text file
bigrams_output_file = "bigrams_output.txt"
with open(bigrams_output_file, "w") as file:
    for bigram, count in bigrams_output:
        file.write(f"{bigram[0]}, {bigram[1]}: {count}\n")

# Print a message indicating successful saving of the output
print("\nOutput has been saved to bigrams_output.txt.")
```
